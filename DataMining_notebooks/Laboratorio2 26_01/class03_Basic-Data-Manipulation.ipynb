### EXERCISE: write a function county_fips_for_state that return all county fips for a given state fips

```python
def county_fips_for_state(state):  #con state intendo il codice del testo
    # deve tornare la lista di tutte le citt√† per uno stato
    result = [] #il risultato sar√† una lista
    
    for k in fips.keys():
        if k[:2].decode('utf-8') == state[:2] and k.decode('utf-8') != state:
            result.append(k) # ogni volta che trovo una key che va bene la
                             # aggiungo alla lista
            
    return result
```

```python
# check for NY
print([(k, fips[k]) for k in list(county_fips_for_state('36000')) ])
```
